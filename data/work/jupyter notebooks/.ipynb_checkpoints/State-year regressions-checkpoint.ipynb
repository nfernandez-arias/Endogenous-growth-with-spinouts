# Data

```r
rm(list = ls())

library(repr)
library(ggplot2)
library(plm)
library(plyr)
library(dplyr)
library(tidyr)
library(data.table)
library(clubSandwich) #for clustered standard errors
library(lmtest)
#library(Hmisc) #for constructing latex tables
library(stargazer) # ibid

data <- fread("../Data/yearStateXrd-RDUserCost-GDP-firmEntryMeasures_stateLevel_compustatAggregation.csv")
names(data)

data <- data %>% rename(normalizationType = IndustryId)

data <- data[normalizationType == 1]

data <- data %>% select(state,year,Firms,firmsTotal,
                    Estabs,estabsTotal,Emp,
                    empTotal,xrd,xrdRaw,GDP,rho_high) 
```

```r
head(data)
```

## Visualizations

### Raw

```r
options(repr.plot.width = 9, repr.plot.height = 5)

ggplot(data,aes(x = xrd, y = Firms, color = state)) + 
    geom_point(size = 0) + 
    geom_smooth(method = lm,size = 0.2, se = FALSE) + 
    labs(x = "R&D (Compustat aggregation)", y = "# of Entering Firms") + 
    ggtitle("Baseline specification: Raw scatterplot")
```

### Normalized variables

```r
options(repr.plot.width = 9, repr.plot.height = 5)

ggplot(data,aes(x = xrd / firmsTotal, y = Firms / firmsTotal, color = state)) + 
    geom_point(size = 0) + 
    geom_smooth(method = lm,size = 0.2, se = FALSE) + 
    labs(x = "R&D (Compustat aggregation)", y = "# of Entering Firms") + 
    ggtitle("Baseline specification: Raw scatterplot")
```

# Regressions

## OLS

### No lags

```r
data[,firmsNorm := Firms / firmsTotal]
data[,estabsNorm := Estabs / estabsTotal]
data[,empNorm := Emp / empTotal]

data[,firmsGDP := Firms / GDP]
data[,estabsGDP := Estabs / GDP]
data[,empGDP := Emp / GDP]

data[,xrdGDP := xrd / GDP]
data[,xrdRawGDP := xrdRaw / GDP]

model1 <- lm(firmsGDP ~ xrdGDP, data = data)
summary(model1)

#model2 <- lm(estabsNorm ~ xrdNorm, data = data)
#summary(model2)

#model3 <- lm(empNorm ~ xrdNorm, data = data)
#summary(model3)
```

### Adding lags

```r
data <- pdata.frame(data, index = c("state","year"))
```

```r
#data[,xrdGDP_lag1 := lag(xrdGDP), by = c("state","year")]

model1 <- plm(firmsGDP ~ xrdGDP + plm::lag(xrdGDP,1) + plm::lag(xrdGDP,2) + plm::lag(xrdGDP,3), data = data,
            model = "pooling")
summary(model1)

#model2 <- plm(estabsNorm ~ xrdNorm + plm::lag(xrdNorm,1) + plm::lag(xrdNorm,2) + plm::lag(xrdNorm,3), data = data)
#summary(model2)

#model3 <- plm(empNorm ~ xrdNorm + plm::lag(xrdNorm,1) + plm::lag(xrdNorm,2) + plm::lag(xrdNorm,3), data = data)
#summary(model3)
```

```r
model1 <- plm(firmsGDP ~ xrdGDP + plm::lag(xrdGDP,1) + plm::lag(xrdGDP,2) + plm::lag(xrdGDP,3), data = data,
              model = "within", effect = "twoways")
summary(model1)

model2 <- plm(estabsGDP )
```
